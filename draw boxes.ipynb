```python
import cv2
import pandas as pd
from PIL import Image
import os
```

```python
curr_path = os.getcwd()
Train_path = os.path.join(curr_path, 'data/tf_wider_train/train_labels.csv')
full_labels = pd.read_csv(Train_path)
```

```python
full_labels.head()
```

```python
def draw_boxes(image_name):
    selected_value = full_labels[full_labels.filename == image_name]
    img = cv2.imread('{}'.format(image_name))
    for index, row in selected_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img
```

```python
Image.fromarray(draw_boxes('00005.jpg'))
```
